### 级联_concat

```python
df1=pd.DataFrame(
    np.random.randint(0,10,size=(5,3)),
    columns=['A','B','C']

)
df2=pd.DataFrame(
    np.random.randint(0,10,size=(3,4)),
    columns=['A','B','C','D']
)
df1,df2
```

```python
pd.concat([df1,df2],axis=0,join='outer')
```

```python
pd.concat([df1,df2],axis=0,join='inner')
```

```python
df3=df1.append(df2)
df3.fillna(axis=1,method='ffill')
```

如果要保留数据的完整性，就要使用outer的方式

### 合并 merge---左右合并，left_on、right_on;内外合并，inner和outer
数据的合并。
级别是表格的拼接
大部分索引一样，一般是做级联
级联不同做合并。

```python
df1 = pd.DataFrame({'employee':['Bob','Jake','Lisa'],
                'group':['Accounting','Engineering','Engineering'],
                })
df1
```

```python
df2 = pd.DataFrame({'employee':['Lisa','Bob','Jake','Jon'],
                'hire_date':[2004,2008,2012,2031],
                })
df2
```

```python
pd.merge(df1,df2,how='outer')
```

```python
pd.merge(df1,df2)#默认how=inner
```

```python
pd.merge(df1,df2,on='employee')#默认how=inner
```

当没有共有列，left_on和right_on分别指定，然后合并

```python
df1 = pd.DataFrame({'employee':['Bobs','Linda','Bill'],
                'group':['Accounting','Product','Marketing'],
               'hire_date':[1998,2017,2018]})
df1
```

```python
df5 = pd.DataFrame({'name':['Lisa','Bobs','Bill'],
                'hire_dates':[1998,2016,2007]})
df5
```

```python
pd.merge(df1,df5,left_on='employee',right_on='name')
```

```python
pd.concat([df1,df5])
```

### 人口数据分析实战

```python
#查看数据
area=pd.read_csv('state-areas.csv')# state  area 
pop=pd.read_csv('state-population.csv')#state/region     ages  year  population
abbrevs=pd.read_csv('state-abbrevs.csv')#state abbreviation
```

```python
area[:3],pop[:3],abbrevs[:3]
```

```python
#合并各州面积
df_1=pd.merge(area,abbrevs,how='outer')
df_1
```

```python
df_1['area (sq. mi)'].isnull().index
```

```python
#合并人口数据和各州简称数据
df1=pd.merge(pop,abbrevs,how='outer',left_on='state/region',right_on='abbreviation')
df1.head()
```

```python
df1.drop('abbreviation',axis=1,replace=True)
```

```python
df1=df1.iloc[:,:-1]
df1.head()
```

```python
from pandas.core.frame import DataFrame
#计算2010全民人口数据
df_pop=df1[df1['year']==2010]
df_pop=df_pop[df_pop['ages']=='total']
df_pop['population'].sum()
```

```python
df_pop
```

```python
#查询操作
df_pop.query('ages=="total" & year==2010')
```

```python
df_area=df_area[df_area['ages']=='total']
df_area=df_area[df_area['year']==2012]
df_area[:3]
```

```python
df_area['dense']=df_area['population']/df_area['area (sq. mi)']
df_area
```
